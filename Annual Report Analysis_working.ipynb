## 2018 Annual Report

```python
filename18 = 'Walmart-AR18.pdf' 

pdfFileObj = open(filename18,'rb')

#The pdfReader variable is a readable object that will be parsed
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)

#discerning the number of pages will allow us to parse through all the pages
num_pages = pdfReader.numPages
count = 0
text18 = ""

while count < num_pages:
    pageObj = pdfReader.getPage(count)
    count +=1
    text18 += pageObj.extractText()
    
#This if statement exists to check if the above library returned words. It's done because PyPDF2 cannot read scanned files.
if text18 != "":
    text18 = text18
#If the above returns as False, we run the OCR library textract to #convert scanned/image based PDF files into text
else:
    text18 = textract.process(fileurl, method='tesseract', language='eng')
```

### 10 Sentence Summary for 2019

```python
word_frequencies_19 = {}  
for word in word_tokenize(text19):  
    if word not in stop_words:
        if word not in word_frequencies_19.keys():
            word_frequencies_19[word] = 1
        else:
            word_frequencies_19[word] += 1
```

```python
maximum_frequency_19 = max(word_frequencies_19.values())

for word in word_frequencies_19.keys():  
    word_frequencies_19[word] = (word_frequencies_19[word]/maximum_frequency_19)
```

```python
import heapq

sentence_list_19 = sent_tokenize(text19)  

sentence_scores_19 = {}  
for sent in sentence_list_19:  
    for word in word_tokenize(sent.lower()):
        if word in word_frequencies_19.keys():
            if len(sent.split(' ')) < 30:
                if sent not in sentence_scores_19.keys():
                    sentence_scores_19[sent] = word_frequencies_19[word]
                else:
                    sentence_scores_19[sent] += word_frequencies_19[word]

summary_sentences_19 = heapq.nlargest(10, sentence_scores_19, key=sentence_scores_19.get)

summary_19 = ' '.join(summary_sentences_19)  
print(summary_19) 
```

```python
import re

#remove numbers
summary_19 =  re.sub(r'\d+', '', summary_19)
summary_19_keywords = [word for word in word_tokenize(summary_19) if not word in stop_words and not word in punctuations]
# summary_19_keywords
```
